```python
import tkinter as tk
from tkinter import ttk
import time
window = tk.Tk()
window.title('猜數字 App')
window.geometry('400x300')
window.configure(background='white')

text = '猜數字小遊戲：\n\n分成簡單&困難模式\n簡單模式沒有次數限制\n困難模式限制5次內完成\n'
header_label = tk.Label(window, text=text)
header_label.pack()

# 隨機製造四個數字為答案
def make_ans():  
    import numpy as np
    L = list('0123456789')
    global ans
    ans = np.random.choice(L, 4, replace = False)
    ans = "".join(ans)
    
# 設成簡單和困難兩種模式
def mode(event):
    global i, t
    if comboxlist.get() == '簡單':
        make_ans() 
        i = 1000000000
        print(ans)
    elif comboxlist.get() == '困難':
        make_ans() 
        i = 5
        print(ans)

# 以下為 hard_frame 群組
hard_frame = tk.Frame(window)
hard_frame.pack(side=tk.TOP) # 向上對齊
hard_label = ttk.Label(hard_frame, text = '難易度')
hard_label.pack(side=tk.LEFT) # 向左對齊
#設置下拉選單
cv = tk.StringVar()
comboxlist = ttk.Combobox(hard_frame, textvariable = cv)
comboxlist.pack(side=tk.TOP)
comboxlist["value"]=("請選擇", "簡單","困難")
comboxlist.current(0)  #選擇第一個  
comboxlist.bind("<<ComboboxSelected>>", mode)  #繫結事件,(下拉列表框被選中時，繫結mode()函式)  
comboxlist.pack() 

# 當按下開始遊戲，跳出另一個視窗
def create():
    # 開始遊戲
    def play():
        # 比對猜的數字和答案
        def guess_number():
            guess = str(guess_entry.get())
            a, b = 0, 0
            for i in range(len(ans)):
                if ans[i] == guess[i]:
                    a += 1
                elif ans[i] in guess:
                    b += 1
            res = '答對了' if a == 4 else f'{a}A{b}B'
            return res
        # 檢查 1.輸入不是文字的東西 2.輸入不是四位數字 3.輸入重複的數字
        def check():
            flag = False
            guess = str(guess_entry.get())
            if guess.isdigit() == False:
                res = '請輸入數字'
                flag = True
            elif len(guess) != 4:
                res = '請輸入4位數字'
                flag = True
            elif sum([guess.count(i) for i in guess]) > 4:
                res = '請不要輸入重複的數字'
                flag = True
            else:
                res = guess_number()
                flag = False
            return flag, res
        
        condition = check()
        # 若檢查有錯誤顯示錯誤的原因
        if condition[0]:
            result_label.configure(text = condition[1])
        # 若檢查沒有錯誤顯示比對答案的結果
        elif condition[0] == False:
            # 若是簡單模式則不顯示次數
            if comboxlist.get() == '簡單':
                result_label.configure(text = condition[1])
            elif comboxlist.get() == '困難':
                global i
                i -= 1
                text = '失敗' if i < 0 else f'{condition[1]}，剩餘{i}次'
                result_label.configure(text =text)

    # 跳出另一個視窗
    window2 = tk.Toplevel()
    window2.title('開始遊戲')
    window2.geometry('400x300')
    window2.configure(background='white')

    # 以下為 guess_frame 群組
    guess_frame = tk.Frame(window2)
    guess_frame.pack(side=tk.TOP)
    guess = tk.StringVar()
    guess_label = tk.Label(guess_frame, text = '你猜的數字')
    guess_label.pack(side=tk.LEFT)
    guess_entry = tk.Entry(guess_frame, textvariable = guess)
    guess_entry.pack(side=tk.LEFT)
    
    result_label = tk.Label(window2)
    result_label.pack()
    
    tk.Button(window2, text='猜測', command=play).pack()

tk.Button(window, text='開始遊戲', command=create).pack()
window.mainloop()
```
